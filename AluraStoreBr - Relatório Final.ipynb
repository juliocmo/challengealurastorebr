```python
import pandas as pd

# URLs dos arquivos CSV
url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

# Leitura dos arquivos
loja1 = pd.read_csv(url1)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

# Calculando o faturamento total por loja
faturamento_loja1 = loja['Preço'].sum()
faturamento_loja2 = loja2['Preço'].sum()
faturamento_loja3 = loja3['Preço'].sum()
faturamento_loja4 = loja4['Preço'].sum()

# Exibindo os resultados
print(f"Faturamento total da Loja 1: R$ {faturamento_loja1:.2f}")
print(f"Faturamento total da Loja 2: R$ {faturamento_loja2:.2f}")
print(f"Faturamento total da Loja 3: R$ {faturamento_loja3:.2f}")
print(f"Faturamento total da Loja 4: R$ {faturamento_loja4:.2f}")
```

# 4. Produtos Mais e Menos Vendidos

```python
import pandas as pd

# URLs dos arquivos CSV
url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

# Leitura dos dados
loja1 = pd.read_csv(url1)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

# Concatenando os dados das lojas
dados_completos = pd.concat([loja1, loja2, loja3, loja4])

# Agrupando os dados por produto e somando as quantidades vendidas
vendas_por_produto = dados_completos.groupby("Produto")["Quantidade de parcelas"].sum()

# Identificando os produtos mais vendidos
produto_mais_vendido = vendas_por_produto.idxmax()
quantidade_mais_vendida = vendas_por_produto.max()

# Identificando os produtos menos vendidos
produto_menos_vendido = vendas_por_produto.idxmin()
quantidade_menos_vendida = vendas_por_produto.min()

# Exibindo os resultados
print(f"Produto mais vendido: {produto_mais_vendido} (Quantidade: {quantidade_mais_vendida})")
print(f"Produto menos vendido: {produto_menos_vendido} (Quantidade: {quantidade_menos_vendida})")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# URLs dos arquivos CSV
url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

# Leitura dos arquivos
loja1 = pd.read_csv(url1)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

# Concatenando os dados das lojas
dados_completos = pd.concat([loja1, loja2, loja3, loja4])

# Calculando a quantidade total vendida por produto
vendas_por_produto = dados_completos.groupby("Produto")["Quantidade de parcelas"].sum()

# Ordenando para identificar os produtos mais e menos vendidos
vendas_ordenadas = vendas_por_produto.sort_values(ascending=False)

# Selecionando os 5 mais vendidos e os 5 menos vendidos
produtos_top5 = vendas_ordenadas.head(5)
produtos_low5 = vendas_ordenadas.tail(5)

# Criando gráfico de pizza para os 5 mais vendidos
plt.figure(figsize=(10, 6))
plt.pie(produtos_top5.values, labels=produtos_top5.index, autopct='%1.1f%%', startangle=140, colors=['green', 'yellow', 'blue', 'orange', 'red'])
plt.title('Produtos Mais Vendidos', fontsize=16)
plt.axis('equal')  # Para manter o formato circular
plt.show()

# Criando gráfico de pizza para os 5 menos vendidos
plt.figure(figsize=(10, 6))
plt.pie(produtos_low5.values, labels=produtos_low5.index, autopct='%1.1f%%', startangle=140, colors=['purple', 'pink', 'cyan', 'magenta', 'brown'])
plt.title('Produtos Menos Vendidos', fontsize=16)
plt.axis('equal')  # Para manter o formato circular
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# URLs dos arquivos CSV
url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

# Leitura dos arquivos
loja1 = pd.read_csv(url1)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

# Calculando o faturamento total por loja
faturamento_loja1 = loja1['Preço'].sum()
faturamento_loja2 = loja2['Preço'].sum()
faturamento_loja3 = loja3['Preço'].sum()
faturamento_loja4 = loja4['Preço'].sum()

# Criando um dataframe com os resultados
faturamento = pd.DataFrame({
    'Loja': ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4'],
    'Faturamento': [faturamento_loja1, faturamento_loja2, faturamento_loja3, faturamento_loja4]
})

# Gerando o gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(faturamento['Loja'], faturamento['Faturamento'], color=['blue', 'green', 'orange', 'red'])
plt.title('Faturamento Total por Loja', fontsize=16)
plt.xlabel('Loja', fontsize=12)
plt.ylabel('Faturamento (R$)', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Análise: A Loja 1 apresentou o maior faturamento, sendo um indicativo de seu desempenho financeiro robusto. Já a Loja 4 apresentou o menor faturamento, evidenciando limitações em potencial de vendas.

```python
import pandas as pd
import matplotlib.pyplot as plt

# URLs dos arquivos CSV
url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

# Leitura dos arquivos
loja1 = pd.read_csv(url1)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

# Calculando a média de avaliação por loja
media_loja1 = loja1['Avaliação da compra'].mean()
media_loja2 = loja2['Avaliação da compra'].mean()
media_loja3 = loja3['Avaliação da compra'].mean()
media_loja4 = loja4['Avaliação da compra'].mean()

# Criando um dataframe com os resultados
avaliacoes = pd.DataFrame({
    'Loja': ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4'],
    'Média de Avaliação': [media_loja1, media_loja2, media_loja3, media_loja4]
})

# Gerando o gráfico de barras
plt.figure(figsize=(8, 6))
plt.bar(avaliacoes['Loja'], avaliacoes['Média de Avaliação'], color=['blue', 'green', 'orange', 'red'])
plt.title('Média de Avaliação por Loja', fontsize=16)
plt.xlabel('Loja', fontsize=12)
plt.ylabel('Média de Avaliação', fontsize=12)
plt.ylim(0, 5.5)  # Definindo limites para o eixo Y
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# URLs dos arquivos CSV
url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

# Leitura dos arquivos
loja1 = pd.read_csv(url1)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

# Calculando o frete médio por loja
frete_medio_loja1 = loja1['Frete'].mean()
frete_medio_loja2 = loja2['Frete'].mean()
frete_medio_loja3 = loja3['Frete'].mean()
frete_medio_loja4 = loja4['Frete'].mean()

# Criando um DataFrame com os resultados
frete_medio = pd.DataFrame({
    'Loja': ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4'],
    'Frete Médio': [frete_medio_loja1, frete_medio_loja2, frete_medio_loja3, frete_medio_loja4]
})

# Gerando o gráfico de barra horizontal
plt.figure(figsize=(8, 6))
plt.barh(frete_medio['Loja'], frete_medio['Frete Médio'], color=['blue', 'green', 'orange', 'red'])
plt.title('Frete Médio por Loja', fontsize=16)
plt.xlabel('Frete Médio (R$)', fontsize=12)
plt.ylabel('Loja', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
